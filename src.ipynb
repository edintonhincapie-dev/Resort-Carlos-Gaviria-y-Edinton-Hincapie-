```python
#DATOS PERSONALES

##NOMBRE
while True:
    nombre = input("Ingrese su nombre: ").strip()

#Quitar espacios internos para validar solo letras
    solo_letras = nombre.replace(" ", "")

#Regla 1: mínimo 3 letras
    if len(solo_letras) < 3:
        print("El nombre debe tener al menos 3 letras (sin contar espacios).")
        continue

#Regla 2: no permitir números ni caracteres especiales
    if not solo_letras.isalpha():
        print("El nombre solo puede contener letras y espacios, no números.")
        continue

#Si todo está bien
    print(f"Nombre válido: {nombre}")
    break

##APELLIDO
while True:
    apellido = input("Ingrese su apellido: ").strip()

#Quitar espacios internos para validar solo letras
    solo_letras = apellido.replace(" ", "")

#Regla 1: mínimo 3 letras reales
    if len(solo_letras) < 3:
        print("El apellido debe tener al menos 3 letras (sin contar espacios).")
        continue

#Regla 2: no permitir números ni caracteres especiales
    if not solo_letras.isalpha():
        print("El apellido solo puede contener letras y espacios, no números.")
        continue

#Si todo está bien
    print(f"Apellido válido: {apellido}")
    break

##IDENTIFICACIÓN
while True:
    documento = input("Ingrese su número de documento: ").strip()

#Regla 1: verificar que solo tenga números
    if not documento.isdigit():
        print("El documento solo puede contener números.")
        continue

#Regla 2: longitud entre 3 y 15 dígitos
    if len(documento) < 3 or len(documento) > 15:
        print("El documento debe tener entre 3 y 15 dígitos.")
        continue

    # Si pasa todas las validaciones
    print(f"Documento válido: {documento}")
    break

##EDAD
while True:
    edad = input("Ingrese su edad: ").strip()

#Verificar que la edad solo tenga números
    if not edad.isdigit():
        print("La edad debe ser un número.")
        continue

    edad = int(edad)

#Clasificación
    if edad < 18:
        print("Es un niño.")
    else:
        print("Es un adulto.")

    break

##MASCOTA
while True:
    mascota = input("¿Lleva mascota? (si / no): ").strip().lower()

    if mascota == "si":
        print("Usted lleva mascota.")
        break
    elif mascota == "no":
        print("Usted NO lleva mascota.")
        break
    else:
        print("Respuesta inválida. Por favor escriba 'si' o 'no'.")
```

```python
#ALISTAMIENTO / LOGÍSTICA
##HABITACIONES
habitaciones = {
    "estandar": {"disponibles": 30, "precio": 20},
    "suite": {"disponibles": 30, "precio": 30},
    "familiar": {"disponibles": 30, "precio": 40}
}

def verificar_disponibilidad_habitacion(habitacion, cantidad):
    tipo = habitacion.lower()

    if tipo not in habitaciones:
        return f"Tipo de habitación '{habitacion}' no existe."

    disponibles = habitaciones[tipo]["disponibles"]
    precio = habitaciones[tipo]["precio"]

    if disponibles >= cantidad:
        return (f"Hay disponibilidad para {tipo}. "
                )
    else:
        return (f"No hay suficientes habitaciones en {tipo}. "
                )


def ocupar_habitacion(habitacion, cantidad):
    tipo = habitacion.lower()

    if tipo in habitaciones and habitaciones[tipo]["disponibles"] >= cantidad:
        habitaciones[tipo]["disponibles"] -= cantidad
        precio_total = habitaciones[tipo]["precio"] * cantidad
        return (f"Reserva exitosa. "
                f"Total a pagar: ${precio_total}")
    else:
        disponibles = habitaciones.get(tipo, {"disponibles": 0})["disponibles"]
        return (f"No se puede reservar.")


#Solicitar datos
habitacion = input("Ingrese la habitación que desea (estandar: $20, suite: $30, familiar: $40): ")
cantidad = int(input("¿Cuántas habitaciones desea reservar?: "))

#Mostrar si hay disponibilidad
print(verificar_disponibilidad_habitacion(habitacion, cantidad))

#Intentar reservar
print(ocupar_habitacion(habitacion, cantidad))

```

```python
#ALISTAMIENTO/LOGISTICA
##TURISMO
cupos = {
    "normal": {"disponibles": 20, "precio": 50},
    "vip": {"disponibles": 20, "precio": 60}
}

def verificar_disponibilidad(plan, cantidad):
    tipo = plan.lower()

    if tipo not in cupos:
        return f"El plan '{plan}' no existe."

    disponibles = cupos[tipo]["disponibles"]
    precio = cupos[tipo]["precio"]

    if disponibles >= cantidad:
        return (f"Hay disponibilidad para el plan {tipo}. "
                f"Precio por persona: ${precio}")
    else:
        return (f"No hay suficientes cupos para el plan {tipo}. "
                f"Precio por persona: ${precio}")

def reservar_plan(plan, cantidad):
    tipo = plan.lower()

    if tipo in cupos and cupos[tipo]["disponibles"] >= cantidad:
        cupos[tipo]["disponibles"] -= cantidad
        total = cupos[tipo]["precio"] * cantidad
        return (f"Reserva exitosa."
                f"Total a pagar: ${total}")
    else:
        return (f"No se puede reservar el plan {tipo}. "
                )

#Solicitar datos al usuario
plan = input("Ingrese el plan que desea (normal: $50, vip: $60): ")
cantidad = int(input("¿Cuántos cupos desea reservar?: "))

#Mostrar disponibilidad
print(verificar_disponibilidad(plan, cantidad))

#Intentar reservar
print(reservar_plan(plan, cantidad))

```

```python
#ALISTAMIENTO/LOGISTICA
##PLAN ALIMENTACION
alimentacion = {
    "1": {"disponibles": 20, "precio": 5},   # Desayuno + Almuerzo
    "2": {"disponibles": 20, "precio": 10}   # Desayuno + Almuerzo + Comida
}

#Verificar disponibilidad
def verificar_disponibilidad(plan, cantidad):
    tipo = plan.lower()

    if tipo not in alimentacion:
        return f"El tipo de plan '{plan}' no existe."

    disponibles = alimentacion[tipo]["disponibles"]
    precio = alimentacion[tipo]["precio"]

    if disponibles >= cantidad:
        return (f"Hay disponibilidad para {cantidad} cupos. "
                f"Precio por persona: ${precio}")
    else:
        return (f"No hay suficientes cupos. "
                )

#Reservar cupos
def reservar_plan(plan, cantidad):
    tipo = plan.lower()

    if tipo not in alimentacion:
        return f"El plan '{plan}' no existe."

    disponibles = alimentacion[tipo]["disponibles"]
    precio = alimentacion[tipo]["precio"]

    if disponibles >= cantidad:
        alimentacion[tipo]["disponibles"] -= cantidad
        total = precio * cantidad
        return (f"Reserva exitosa de {cantidad} cupos. "
                f"Total a pagar: ${total}")
    else:
        return (f"No se puede reservar. Cupos insuficientes para '{plan}'. "
                )



#Solicitar plan al usuario
plan = input("Ingrese el plan que desea (1 (Desayuno + almuerzo) o 2 (Desayuno + Almuerzo + Comida)): ")
cantidad = int(input("¿Cuántos cupos desea?: "))

#Mostrar disponibilidad
print(verificar_disponibilidad(plan, cantidad))

#Intentar reservar
print(reservar_plan(plan, cantidad))

```
